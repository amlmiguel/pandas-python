# Converter String para n√∫mero

```
df = pd.DataFrame({'col_one':['1.1', '2.2', '3.3'],
                   'col_two':['4.4', '5.5', '6.6'],
                   'col_three':['7.7', '8.8', '-']})
df
```

```
df.dtypes
```

```
df.astype({'col_one':'float','col_two':'float'}).dtypes
```

Entretanto, se tentarmos incluir a tereira coluna nesse esquema ter√≠amos um erro. Isso porque a terceira coluna na terceria linha √© um h√≠fen, e o pandas n√£o sabe como interpreta-lo

Ao √≠nves disso, podemos usar to_numeric() e explicitar apra converter um caracter invalido para NaN

```
pd.to_numeric(df.col_three,errors='coerce')
```

```
pd.to_numeric(df.col_three, errors='coerce').fillna(0)
```

Para aplicar isso em todo data frame usamos o apply()

```
df = df.apply(pd.to_numeric, errors='coerce').fillna(0)
df
```

```
df.dtypes
```
